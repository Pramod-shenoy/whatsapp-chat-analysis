```python
import re
import pandas as pd
```

```python
f = open('WhatsApp Chat with Chips  squareğŸ˜ğŸ˜µ.txt', 'r', encoding='utf-8')
data = f.read()
```

```python
print(data)
```

```python
data = data.replace('\u202f', ' ')
```

```python
pattern = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s[APMapm]{2}\s-\s'
```

```python
messages = re.split(pattern, data)[1:]
messages
```

```python
dates = re.findall(pattern, data)
dates
```

```python
import pandas as pd
df = pd.DataFrame({'user_message': messages, 'message_date': dates})
df['message_date'] = df['message_date'].str.replace(r' -\s*$', '', regex=True)
df['message_date'] = pd.to_datetime(df['message_date'], format='%m/%d/%y, %I:%M %p')
df.rename(columns={'message_date': 'date'}, inplace=True)
df.head()
```

```python
df.shape
```

```python
import re
import pandas as pd

# Check the column names to see if 'user_message' exists
if 'user_message' not in df.columns:
    print("The column 'user_message' is not present.")
else:
    users = []
    messages = []

    # Iterate over the 'user_message' column to separate users and messages
    for message in df['user_message']:
        entry = re.split(r'([\w\W]+?):\s', message)  # Regex to split the message by user name
        
        if len(entry) > 1:  # If there's a user in the message
            users.append(entry[1])
            messages.append(entry[2])
        else:  # For group notifications (no user name)
            users.append('group_notification')
            messages.append(entry[0])

    # Add the new 'user' and 'message' columns to the DataFrame
    df['user'] = users
    df['message'] = messages

    # Drop the original 'user_message' column
    df.drop(columns=['user_message'], inplace=True)

    # Display the first few rows
    df.head()
```

```python
df['only_date'] = df['date'].dt.date
df['year'] = df['date'].dt.year
df['month_num'] = df['date'].dt.month
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
df['day_name'] = df['date'].dt.day_name()
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
```
